```python
df=pd.read_csv('data.csv')
df.tail()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=1)            
df.tail(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le_y=LabelEncoder()
le_y.fit_transform(df.iloc[:,1].values)               
```

```python
df.iloc[:,1]=le_y.fit_transform(df.iloc[:,1].values)                 
```

```python
df.tail()
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(df.iloc[:,1:12].corr(),annot=True, cmap="Spectral", fmt='.0%')
plt.show()
#cubehelix
```

```python

x=df.iloc[:,2:31].values              
y=df.iloc[:,1].values                 
from sklearn.model_selection import train_test_split                                     
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=0, stratify=y)       
```

```python
# Scaling of data
from sklearn.preprocessing import StandardScaler                                
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
import tensorflow as tf
from sklearn.metrics import  r2_score
```

```python
x.shape[1]
```

```python

model = tf.keras.Sequential(
    [tf.keras.layers.Dense(5, activation="relu", input_shape=(x.shape[1],)),
     tf.keras.layers.Dense(8, activation="relu"),
     tf.keras.layers.Dense(1, activation="sigmoid")]
)
```

```python
model.summary()
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics='accuracy')
```

```python
trained_model = model.fit(x_train, y_train, epochs=100, batch_size=20)
```

```python
plt.plot(trained_model.history["loss"])
plt.show()
```

```python
# testing
y_hat = model.predict(x_test)
```

```python
y_hat1 = np.where(y_hat >= 0.5, 1, 0)
```

```python
y_hat1.flatten()
```

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
from sklearn.metrics import roc_auc_score
#roc_auc_score(y_test, y_hat1)
print(color.UNDERLINE, color.BOLD, color.RED + str(roc_auc_score(y_test, y_hat1) ) +color.END)
```
